***Below is the list of all the files I downloaded from inside airbnb. Each file is a scraping of the Airbnb website for listings in SF for the specified month and year***

```python
fileList = ['nov-2015.csv', 'may-2016.csv', 'nov-2016.csv', 'jan-2017.csv', 'feb-2017.csv', 'mar-2017.csv', 'apr-2017.csv', 'may-2017.csv', 'jun-2017.csv', 'jul-2017.csv', 'aug-2017.csv', 'sep-2017.csv', 'oct-2017.csv', 'nov-2017.csv', 'dec-2017.csv', 'may-2018.csv', 'nov-2018.csv', 'may-2019.csv']
```

```python
finalData = pd.read_csv("may-2015.csv", dtype={'id' : float, 'latitude': float, 'longitude': float, 'price' : float, 'monthly_price' : float, 'minimum_nights' : float, "maximum_nights" : float,'neighbourhood_cleansed' : str,'property_type' : str, 'room_type' : str, 'license' : str})
finalData = finalData[['id', 'latitude', 'longitude', 'price', 'monthly_price', 'minimum_nights', "maximum_nights",'neighbourhood_cleansed', 'property_type', 'room_type', 'license']]
finalData['year'] = 2015
finalData['month'] = "may"
```

```python
finalData
```

```python
low_memory = False
for file in fileList:
    data = pd.read_csv(file, dtype={'id' : float, 'latitude': float, 'longitude': float, 'price' : float, 'monthly_price' : float, 'minimum_nights' : float, "maximum_nights" : float,'neighbourhood_cleansed' : str,'property_type' : str, 'room_type' : str, 'license' : str})
    data = data[['id', 'latitude', 'longitude', 'price', 'monthly_price', 'minimum_nights', "maximum_nights",'neighbourhood_cleansed','property_type', 'room_type', 'license']]
    monthAndYear = file[:-4]
    month = monthAndYear[:3]
    year = monthAndYear[-4:]
    data['year'] = year
    data['month'] = month
    finalData = pd.concat([finalData, data], axis = 0)
```

```python
finalData
```

***Above, I have concatenated each of the datasets for the individual month/year combos together. In doing so, I added a two this final dataset, one for the year and one for the month to which the specific instance came from.***

***Now below, I work on adding the SF average rental price data to this combined dataset***

```python
priceData = pd.read_excel('SF-Rental-Price-Data-Final.xlsx')
```

```python
dates = []
for i in range(len(priceData)):
    dates.append(str(priceData['Month'][i]))
years = []
months = []
monthNumbers = {1: 'jan', 2: 'feb', 3: 'mar', 4: 'apr', 5: 'may', 6: 'jun', 7:'jul', 8:'aug', 9:'sep', 10:'oct',11:'nov',12:'dec'}
for i in dates:
    year = int(i[:4])
    month = monthNumbers[int(i[5:7])]
    years.append(year)
    months.append(month)
priceData['year'] = years
priceData['month'] = months
```

***Above I have taken note of the month and year for which each price is given. I only want the prices for the month/year combinations that are present in the combined dataset made above***

```python
averageRentalPrices = []
months = list(finalData['month'])
years = list(finalData['year'])
for i in range(len(finalData)):
    month = months[i]
    year = int(years[i])
    match = priceData[(priceData['year'] == year) & (priceData['month'] == month)]
    price = match["All Beds"].sum()
    averageRentalPrices.append(price)
```

```python
finalData['comparable_price'] = averageRentalPrices
```

***Above, I have matched each instance in the combined dataset with the appropriate price in priceData, and then appended these results to the combined dataset***
